```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('adult.csv')
df.head()
```

```python
print(df.columns)
```

```python
# Supress marital.status, relationship, race, native.country
df = df.drop(['marital.status', 'relationship', 'race', 'native.country'], axis=1)
df.head()
```

```python
print(df['occupation'].value_counts())
```

```python
# Data Generalization

df['occupation'] = df['occupation'].replace(['Prof-specialty', 'Exec-managerial', 'Adm-clerical', 'Tech-support'], 'Professional and Administrative')
print(df['occupation'].value_counts())

df['occupation'] = df['occupation'].replace(['Craft-repair', 'Machine-op-inspct', 'Transport-moving', 'Handlers-cleaners', 'Farming-fishing'], 'Manual and Field Work')
print(df['occupation'].value_counts())

df['occupation'] = df['occupation'].replace(['Sales', 'Other-service', 'Protective-serv', 'Priv-house-serv'], 'Sales and Services')
print(df['occupation'].value_counts())

df['occupation'] = df['occupation'].replace(['Armed-Forces', '?'], 'Other')
print(df['occupation'].value_counts())
```
